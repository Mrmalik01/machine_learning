```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

%matplotlib inline
matplotlib.style.use('ggplot')

# create data samples 

df = pd.DataFrame({
    "x1" : np.random.normal(-10, 5, 50),
    "x2" : np.random.normal(30, 10, 50),
    "x3" : np.random.normal(0, 2, 50)
})

display(df)

# Scaling Standardisation

scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=['x1', 'x2', 'x3'])
display(scaled_df)

# Ploating the values 

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,5))
ax1.set_title("Before scaling")
sns.kdeplot(df['x1'], ax = ax1)
sns.kdeplot(df['x2'], ax = ax1)
sns.kdeplot(df['x3'], ax = ax1)

ax2.set_title("After scaling")
sns.kdeplot(scaled_df['x1'], ax = ax2)
sns.kdeplot(scaled_df['x2'], ax = ax2)
sns.kdeplot(scaled_df['x3'], ax = ax2)

plt.plot()
print("fig value {}".format(fig))
print("ax1 value {}".format(ax1))
print("ax2 value {}".format(ax2))
```

```python
# Normalisation


# Creating data samples

df = pd.DataFrame({
    # positive skew
    'x1': np.random.chisquare(8, 1000),
    # negative skew 
    'x2': np.random.beta(8, 2, 1000) * 40,
    # no skew
    'x3': np.random.normal(50, 3, 1000)
})

# MinMax scaling - Normalisation

scaler = preprocessing.MinMaxScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=['x1','x2','x3'])

# ploting

figure, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,5))

ax1.set_title("Before scaling")
sns.kdeplot(df['x1'], ax=ax1)
sns.kdeplot(df['x2'], ax=ax1)
sns.kdeplot(df['x3'], ax=ax1)

ax2.set_title("After scaling")
sns.kdeplot(scaled_df['x1'], ax=ax2)
sns.kdeplot(scaled_df['x2'], ax=ax2)
sns.kdeplot(scaled_df['x3'], ax=ax2)

plt.show()
```
